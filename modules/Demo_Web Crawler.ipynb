```python
import pandas
import pandas_datareader as pdr
from datetime import datetime

# sd = datetime(2017, 8, 30)
# ed = datetime(2019, 8, 30)
sd = datetime(2019, 9, 27)
ed = datetime(2020, 10, 1)

vix = pdr.get_data_yahoo(symbols='^vix', start=sd, end=ed)["Adj Close"]
spx = pdr.get_data_yahoo(symbols='^SP500TR', start=sd, end=ed)["Adj Close"]
vfx = pdr.get_data_yahoo(symbols='VFINX', start=sd, end=ed)["Adj Close"]
vbx = pdr.get_data_yahoo(symbols='VBMFX', start=sd, end=ed)["Adj Close"]
vmt = pdr.get_data_yahoo(symbols='VMOT', start=sd, end=ed)["Adj Close"]
rwm = pdr.get_data_yahoo(symbols='RWM', start=sd, end=ed)["Adj Close"]
dog = pdr.get_data_yahoo(symbols='DOG', start=sd, end=ed)["Adj Close"]
psh = pdr.get_data_yahoo(symbols='SH', start=sd, end=ed)["Adj Close"]
# gld = pdr.get_data_yahoo(symbols='GLD', start=sd, end=ed)["Adj Close"]
# uso = pdr.get_data_yahoo(symbols='USO', start=sd, end=ed)["Adj Close"]
# CNY_X= pdr.get_data_yahoo(symbols='CNY=X', start=sd, end=ed)["Adj Close"] 
# BTCUSD_X= pdr.get_data_yahoo(symbols='BTCUSD=X', start=sd, end=ed)["Adj Close"] 
```

```python
bnp = pandas.concat([vfx,vix,vbx,vmt,rwm,dog,psh,spx], axis=1)
bnp.columns=['vfx','vix','vbx','vmt','rwm','dog','psh','spx']
bnp = bnp.sort_values(by="Date",ascending=True)
bnp
bnp.isnull().sum()
bnp
```

```python
bnp.to_csv('PRED2.csv')
```
