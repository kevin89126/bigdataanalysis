```python
from LSTM_FINAL import read_csv
import matplotlib.pyplot as plt
```

```python
res = read_csv('LSTM_RES.csv','/nfs/Workspace/')
```

```python
res = res[:-1]
mean = res['MEAN']
up_bond = res['UP_BOND']
low_bond = res['LOW_BOND']
pred_rate = res['PRED_RATE']
real_rate = res['REAL_RATE']
date = res['REAL_DATE'].to_list()
```

```python
plt.figure(figsize=(20,10))
plt.plot(date,real_rate, color = 'red', label = 'Real')
plt.plot(date,pred_rate, color = 'blue', label = 'Predict')
plt.plot(date,up_bond, color = 'green', label = 'Up Bond')
plt.plot(date,low_bond, color = 'yellow', label = 'Low Bond')
plt.plot(date,mean, color = 'black', label = 'Mean')
plt.title('Return Rate with Time')
plt.xlabel('Time')
plt.ylabel('Retrun Rate')
plt.legend()
plt.show()
```
