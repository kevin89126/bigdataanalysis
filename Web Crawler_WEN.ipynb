</div>

```python
import pandas
import pandas_datareader as pdr
import numpy as np
from datetime import datetime

sd = datetime(2020, 12, 25)
ed = datetime(2020, 12, 25)

vfx = pdr.get_data_yahoo(symbols='VFINX', start=sd, end=ed)["Adj Close"]
#vix = pdr.get_data_yahoo(symbols='^vix', start=sd, end=ed)["Adj Close"]
spx = pdr.get_data_yahoo(symbols='^SP500TR', start=sd, end=ed)["Adj Close"]
vbx = pdr.get_data_yahoo(symbols='VBMFX', start=sd, end=ed)["Adj Close"]
vmt = pdr.get_data_yahoo(symbols='VMOT', start=sd, end=ed)["Adj Close"]

rwm = pdr.get_data_yahoo(symbols='RWM', start=sd, end=ed)["Adj Close"]
dog = pdr.get_data_yahoo(symbols='DOG', start=sd, end=ed)["Adj Close"]
psh = pdr.get_data_yahoo(symbols='SH', start=sd, end=ed)["Adj Close"]
```

```python
bnp = pandas.concat([vfx,vbx,vmt,rwm,dog,psh,spx], axis=1)
print(bnp)
bnp = bnp.sort_values(by="Date",ascending=True)
bnp.head(80)
```

```python
bnp.to_csv('ONCE_PRED.csv')
```

<div class="alert alert-block alert-info">
<b>News</b> : News Crawler (LTN)
</div>

```python
import requests
res = requests.get('https://news.ltn.com.tw/list/breakingnews')
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(res.text,'lxml')
```

```python
newsary = []
for li in soup.select('ul.list li'):
  #print(li)
  title = li.select_one('.title').text
  dt = li.select_one('.time').text
  link = li.select_one('a').get('href')
  newsary.append({'title':title, 'time': dt, 'link':link})
```

```python
import pandas
newsdf = pandas.DataFrame(newsary)
newsdf.head()
```

```python
import requests
res = requests.get('https://news.ltn.com.tw/ajax/breakingnews/all/4')
jd = res.json()
import pandas
df = pandas.DataFrame(jd['data'].values())
import time
newsary = []
for i in range(2,5):
  res = requests.get('https://news.ltn.com.tw/ajax/breakingnews/all/{}'.format(i))
  time.sleep(2)
  #print(res.json())
  jd = res.json()
  df = pandas.DataFrame(jd['data'].values())
  newsary.append(df)
```

```python
newsdf = pandas.concat(newsary)
```

```python
import requests
payload = {
  'encodeURIComponent': '1',
  'step': '1',
  'firstin': '1',
  'TYPEK': 'sii',
  'year': '109',
  'season': '01'    
}

url = 'https://mops.twse.com.tw/mops/web/ajax_t163sb19'

res = requests.post(url, data = payload)
```

```python
import pandas
dfs = pandas.read_html(res.text)
df = dfs[0]
df[df['產業別'] != '加權平均數']
```
